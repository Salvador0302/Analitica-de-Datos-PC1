```python
# Eliminar columnas irrelevantes
cols_to_drop = [
    'id_tipo_hecho','id_materia_hecho','id_subtipo_hecho','id_modalidad_hecho',
    'id_dpto_hecho','id_prov_hecho','id_dist_hecho','lat','lon',
    'ubigeo_cia_registro','ubigeo_cia_hecho','id_dgc','id_comisaria_registro',
    'cod_uni_hecho','cod_cpnp_hecho','cod_macroregpol_hecho','cod_regpol_hecho',
    'cod_divpol_divopus_hecho','objectid'
]

for c in cols_to_drop:
    if c in df.columns:
        df.drop(columns=c, inplace=True)

print(f"Columnas restantes: {len(df.columns)}")
df.head()
```

```python
# Convertir fecha_hora_hecho a datetime y eliminar mes_hecho, dia_hecho, ano_hecho
df['fecha_hora_hecho'] = pd.to_datetime(df['fecha_hora_hecho'], errors='coerce')
for c in ['mes_hecho','dia_hecho','ano_hecho']:
    if c in df.columns:
        df.drop(columns=c, inplace=True)
print('dtype fecha_hora_hecho ->', df['fecha_hora_hecho'].dtype)
print('Columnas totales ->', len(df.columns))
df.head()
```

```python
#Eliminar columnas administrativas irrelevantes
cols_to_drop = ["fecha_hora_registro_hecho", "turno_hecho_cod",
                "estado", "fuente", "observacion", "estado_coord", "tiene_coordenada"]

for c in cols_to_drop:
    if c in df.columns:
        df.drop(columns=c, inplace=True)
        
print('Columnas totales ->', len(df.columns))
df.head()
```

### Imputación de valores faltantes (categorías)
- Columnas: `tipo_via_hecho`, `divpol_divopus_hecho`.
- Decisión: completar con "Desconocido" (no se interpola ni media/mediana porque son categóricas; la moda puede sesgar hacia la clase mayoritaria).
- Beneficio: preserva filas, evita sesgo y mantiene semántica de "valor no observado".

```python
cols = ['tipo_via_hecho','divpol_divopus_hecho','direccion_hecho','modalidad_hecho','subtipo_hecho','tipo_hecho','comisaria_hecho','regionpol_hecho','macroregpol_hecho']; df[cols]=df[cols].fillna('Desconocido'); df[cols].isna().sum()
```

## Análisis univariado
Primero variables categóricas (conteos) y luego numéricas (histogramas).

```python
import matplotlib.pyplot as plt 
import seaborn as sns

fig, ax = plt.subplots(1, 3, figsize=(18, 6))
items = [
    ('departamento_hecho', 'Departamentos', None),
    ('provincia_hecho', 'Provincias (Top 30)', 30),
    ('distrito_hecho', 'Distritos (Top 30)', 30)
]
for axes, (col, title, top) in zip(ax, items):
    s = df[col].value_counts(dropna=False)
    if top: s = s.head(top)
    sns.barplot(ax=axes, x=s.values, y=s.index, orient='h')
    axes.set_title(title); axes.set_xlabel('Conteo'); axes.set_ylabel(col)

plt.tight_layout(); plt.show()
```

```python
fig, axis = plt.subplots(2, 2, figsize = (10, 7), gridspec_kw={'height_ratios': [6, 1]})

# Crear una figura múltiple con histogramas y diagramas de caja
sns.histplot(ax = axis[0, 0], data = df, x = 'lat_hecho').set(xlabel = None)
sns.boxplot(ax = axis[1, 0], data = df, x = 'lat_hecho')
sns.histplot(ax = axis[0, 1], data = df, x = 'long_hecho').set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[1, 1], data = df, x = 'long_hecho')

plt.tight_layout(); plt.show()
```
